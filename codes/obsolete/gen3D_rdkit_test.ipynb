```python
from rdkit import Chem
from rdkit.Chem import AllChem
import numpy as np
from rdkit.Chem import PandasTools
import pandas as pd
```

```python
m_smiles ='C[C@]12CC[C@H]3[C@@H](CCC4=CC(=O)CC[C@]34C)[C@@H]1CC[C@@H]2O'
#[H][C@@]12CC[C@H](C(C)=O)[C@@]1(C)CC[C@@]1([H])[C@@]2([H])CCC2=CC(=O)CC[C@]12C  progesterone
#C[C@]12CC[C@H]3[C@@H](CCC4=CC(=O)CC[C@]34C)[C@@H]1CC[C@@H]2O testosterone
#[2H]C1=C([2H])C2=C3C(=C1[2H])C1=C([2H])C([2H])=C([2H])C4=C1C(=C([2H])C([2H])=C4[2H])C3=C([2H])C([2H])=C2[2H] perylened12

mol = Chem.MolFromSmiles(m_smiles)
mol = Chem.AddHs(mol)
mol.GetPropNames()
for prop in mol.GetPropNames():
    print(str(prop))

symbols = list(set([a.GetSymbol() for a in mol.GetAtoms()]))
num_atoms = len(mol.GetAtoms())
```
